```python
Mystore = {
    's': {
        'kurti': {
            'shortkurti': {'Quantity': 200, 'Price': 2000},
            'Longkurti': {'Quantity': 200, 'Price': 2250},
            'Anarkali kurti': {'Quantity': 100, 'Price': 3000},
            'jacket style kurti': {'Quantity': 100, 'Price': 1500}
        },
        'Jeans': {
            'Bootcut jeans': {'Quantity': 100, 'Price': 1000},
            'Wide Jeans': {'Quantity': 200, 'Price': 1500},
            'Straight Jeans': {'Quantity': 200, 'Price': 1000},
            'Mom Jeans': {'Quantity': 150, 'Price': 2500}
        }
    },
    'm': {
        'kurti': {
            'shortkurti': {'Quantity': 200, 'Price': 2000},
            'Longkurti': {'Quantity': 200, 'Price': 2250},
            'Anarkali kurti': {'Quantity': 100, 'Price': 3000},
            'jacket style kurti': {'Quantity': 100, 'Price': 1500}
        },
        'Jeans': {
            'Bootcut jeans': {'Quantity': 100, 'Price': 1000},
            'Wide Jeans': {'Quantity': 200, 'Price': 1500},
            'Straight Jeans': {'Quantity': 200, 'Price': 1000},
            'Mom Jeans': {'Quantity': 150, 'Price': 2500}
        }
    },
    'l': {
        'kurti': {
            'shortkurti': {'Quantity': 200, 'Price': 2000},
            'Longkurti': {'Quantity': 200, 'Price': 2250},
            'Anarkali kurti': {'Quantity': 100, 'Price': 3000},
            'jacket style kurti': {'Quantity': 100, 'Price': 1500}
        },
        'Jeans': {
            'Bootcut jeans': {'Quantity': 100, 'Price': 1000},
            'Wide Jeans': {'Quantity': 200, 'Price': 1500},
            'Straight Jeans': {'Quantity': 200, 'Price': 1000},
            'Mom Jeans': {'Quantity': 150, 'Price': 2500}
        }
    },
    'xl': {
        'kurti': {
            'shortkurti': {'Quantity': 200, 'Price': 2000},
            'Longkurti': {'Quantity': 200, 'Price': 2250},
            'Anarkali kurti': {'Quantity': 100, 'Price': 3000},
            'jacket style kurti': {'Quantity': 100, 'Price': 1500}
        },
        'Jeans': {
            'Bootcut jeans': {'Quantity': 100, 'Price': 1000},
            'Wide Jeans': {'Quantity': 200, 'Price': 1500},
            'Straight Jeans': {'Quantity': 200, 'Price': 1000},
            'Mom Jeans': {'Quantity': 150, 'Price': 2500}
        }
    },
    'xxl': {
        'kurti': {
            'shortkurti': {'Quantity': 200, 'Price': 2000},
            'Longkurti': {'Quantity': 200, 'Price': 2250},
            'Anarkali kurti': {'Quantity': 100, 'Price': 3000},
            'jacket style kurti': {'Quantity': 100, 'Price': 1500}
        },
        'Jeans': {
            'Bootcut jeans': {'Quantity': 100, 'Price': 1000},
            'Wide Jeans': {'Quantity': 200, 'Price': 1500},
            'Straight Jeans': {'Quantity': 200, 'Price': 1000},
            'Mom Jeans': {'Quantity': 150, 'Price': 2500}
        }
    }
}


cart = []

def display_store():
    print("\nAvailable Categories and Clothing Items:")
    for size, categories in Mystore.items():
        print(f"\nSize: {size.upper()}")
        for category, subcategories in categories.items():
            print(f"  Category: {category.capitalize()}")
            for item_name, item_info in subcategories.items():
                print(f"    - {item_name} - Price: {item_info['Price']} Rupees - Stock: {item_info['Quantity']}")

def select_size():
    print("\nAvailable Sizes:")
    for idx, size in enumerate(Mystore.keys(), 1):
        print(f"{idx}. {size.upper()}")
    size_choice = int(input("\nSelect a size by number (0 to cancel): "))
    if size_choice == 0:
        return None
    return list(Mystore.keys())[size_choice - 1]

def select_category(size):
    categories = list(Mystore[size].keys())
    print("\nAvailable Categories:")
    for idx, cat in enumerate(categories, 1):
        print(f"{idx}. {cat.capitalize()}")
    choice = int(input("Select a category by number (0 to cancel): "))
    if choice == 0:
        return None
    return categories[choice - 1]

def select_item(size, category):
    items = list(Mystore[size][category].items())
    for idx, (name, info) in enumerate(items, 1):
        print(f"{idx}. {name} - {info['Price']} Rupees - Stock: {info['Quantity']}")
    choice = int(input("Enter item number (0 to cancel): "))
    if choice == 0:
        return None
    return items[choice - 1]

def add_to_cart():
    selected_size = select_size()
    if not selected_size:
        return

    selected_category = select_category(selected_size)
    if not selected_category:
        return

    item = select_item(selected_size, selected_category)
    if not item:
        return

    item_name, item_info = item
    quantity = int(input(f"How many {item_name} would you like to add? "))

    if quantity <= item_info['Quantity']:
        cart.append({"item": item_name, "quantity": quantity, "price": item_info['Price']})
        Mystore[selected_size][selected_category][item_name]['Quantity'] -= quantity
        print(f"Added {quantity} x {item_name} to your cart.")
    else:
        print(f"Only {item_info['Quantity']} in stock.")

def view_cart():
    if not cart:
        print("\nYour cart is empty.")
        return
    print("\nYour Cart:")
    total = 0
    for item in cart:
        subtotal = item['quantity'] * item['price']
        total += subtotal
        print(f"{item['item']} - {item['quantity']} x {item['price']} = {subtotal} Rupees")
    print(f"\nTotal Price: {total} Rupees")

def checkout():
    if not cart:
        print("Cart is empty.")
        return

    total = sum(item['quantity'] * item['price'] for item in cart)
    print(f"\nTotal Price: {total} Rupees")

    delivery_option = input("Delivery or Pickup? (delivery/pickup): ").lower()
    name = input("Enter your name: ")
    contact = input("Phone number: ")

    if delivery_option == "delivery":
        address = input("Enter your delivery address: ")
        print(f"\nOrder Summary:\nName: {name}\nAddress: {address}\nContact: {contact}")
        print("Your items will be delivered soon!")
    elif delivery_option == "pickup":
        print(f"\nOrder Summary:\nName: {name}\nContact: {contact}")
        print("Your items are ready for pickup.")
    else:
        print("Invalid delivery option selected.")


while True:
    action = input("\nWhat would you like to do? (add/view/checkout/exit): ").lower()

    if action == "add":
        display_store()  
        add_to_cart()
    elif action == "view":
        view_cart()
    elif action == "checkout":
        checkout()
        break
    elif action == "exit":
        print("Thank you for visiting!")
        break
    else:
        print("Invalid input.")

```
